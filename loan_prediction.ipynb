```python
import pandas as pd
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("/content/drive/MyDrive/loan.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
data.columns
```

```python
encoder.fit(data['Gender'])
```

```python
data['Gender']=encoder.transform(data['Gender'])
```

```python
data.head()
```

```python
data.drop('gender',axis=1,inplace=True)
```

```python
data.shape
```

```python
li=['Married','Education','Self_Employed','Dependents','Property_Area','Loan_Status']
```

```python
for item in li:
  encoder.fit(data[item])
  data[item]=encoder.transform(data[item])
```

```python
data.head()
```

```python
data['Dependents'].value_counts()
```

```python
x=data.drop(columns=['Loan_ID','Loan_Status'],axis=1)
```

```python
y=data['Loan_Status']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(pred,y_test)
```

```python
print(accuracy*100)
```
